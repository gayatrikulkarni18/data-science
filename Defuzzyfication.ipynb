```python
import matplotlib.pyplot as plt

#Membership function for "Speed" linguistic variable
Î¼Slow = 0.2
Î¼Moderate = 0.7
Î¼Fast = 0.9

#Lambda (Î») value
Î» = 0.6
# Initialize an empty list to store the linguistic in the Lambda-Cut
lambda_cut = []

if Î¼Slow <= Î»:
    lambda_cut.append("Slow")

if Î¼Moderate > Î»:
    lambda_cut.append("Moderate")
    
#Lambda-Cut for "Fast"
if Î¼Fast > Î»:
    lambda_cut.append("Fast")

#Visualize the membership function
x = [0, 1, 2, 3] #X-axis values
y = [0, Î¼Slow, Î¼Moderate, Î¼Fast] #Membership values

# Plot the membership function
plt.plot(x,y, label="Membership Function")
plt.fill_between([0, 1, 2, 3], [0, 0, 0, 0],[0, Î¼Slow, Î¼Moderate, Î¼Fast],  alpha=0.2, color='blue', label=f'Î» = {Î»}Cut')
plt.xlabel("Speed")
plt.ylabel("Membership Function with Lambda-cut")
plt.legend()
plt.grid(True)
                 
#Display the Lambda-Cut result
if not lambda_cut:
                result = "No linguistic variables in Lambda-Cut"
else:
    result =" ," .join(lambda_cut)
print(f"Lambda-Cut for Î» = {Î»}:{result}")
#Show the plot
plt.show()
print("Gayatri Kulkarni-53004230002")
```
